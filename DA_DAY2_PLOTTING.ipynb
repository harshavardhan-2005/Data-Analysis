```python
import numpy as np
import matplotlib.pyplot as plt

runs = np.array([100, 50, 91, 78, 89, 25, 34, 19, 9, 10])
w = np.array([1, 0, 2, 0, 3, 7, 8, 9, 7, 5])
plt.scatter(runs, w, color="red",marker="s")
plt.show()
```

```python
#plot the scores of two students in 5 different shapes
#sub in x axis
#marks in y axis
s1=[99,71,63,73,100]
s2=[57,92,74,75,60]
sub=["sanskrit","English","math","math1","physics"]
plt.plot(sub,s1,label="student1",color="k")
plt.plot(sub,s2,label="student2",color="m")
plt.title("MARKS VS SUBJECTS")
plt.legend()
```

```python
#plot the scores of two students in 5 different shapes IN SUB PLOTS
#sub in x axis
#marks in y axis
s1=[99,71,63,73,100]
s2=[57,92,74,75,60]
sub=["sanskrit","English","math","math1","physics"]
plt.subplot(1,2,1)
plt.plot(sub,s1,label="student1",color="k")
plt.title("MARKS VS SUBJECTS")
plt.legend()
plt.subplot(1,2,2)
plt.plot(sub,s2,label="student2",color="m")
plt.title("MARKS VS SUBJECTS")
plt.legend()
```

```python
years=np.array(['2019-20','2020-21','2021-22','2022-23'])
a=np.array([230,560,780,127,128])
b=np.array([200,160,270,127,400])
p1=np.diff(a)
p2=np.diff(b)
plt.subplot(2,1,1)
plt.title("profits of a company")
plt.bar(years,p1,label="company A",color="g")
plt.legend()
plt.subplot(2,1,2)
plt.title("profits of b company")
plt.bar(years,p2,label="company b",color="orange")
plt.legend()
```

```python
a=np.array([25,60,5,10])
labe=["AIML","PYTHON","PANDAS","NUMPY"]
explo=[0.2,0,0,0]
colors=["yellow","orange","blue","m"]
plt.pie(a,labels=labe, explode=explo, startangle = 45, shadow=True, colors=colors)
plt.legend()
plt.show()
```
